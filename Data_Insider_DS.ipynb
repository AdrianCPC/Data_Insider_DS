```python
api_data = 'QOpPxC_54Rcq_z48oY2PfNTWANdJlAJK'
```

```python
#Nombres acciones empresas
name_companies = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'FB']

#fechas
data_initial = '2022-01-01'
data_end = '2022-12-31'
```

```python
#Solicitud de API y Dataframes

data_companies = []

for company in name_companies:
  url = f"https://api.polygon.io/v2/aggs/ticker/{company}/range/1/day/{data_initial}/{data_end}?apiKey={api_data}"
  response = requests.get(url)
  data = response.json()

  df = pd.DataFrame(data['results'])
  df['company'] = company

  data_companies.append(df)

  #Union de dataframes

final_dataset = pd.concat(data_companies)
```

```python
final_dataset.head(-1)
```

```python
#Guardado de archivos en csv

final_dataset.to_csv('historico_acciones.csv', index=False)
```
